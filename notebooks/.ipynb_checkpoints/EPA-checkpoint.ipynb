```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fname = '../original_data/Sales and Order by Month.csv'
df = pd.read_csv(fname)
df = df.replace(',','',regex=True)
df.head()
```

```python
df['Ordered Product Sales'] = df['Ordered Product Sales'].str.replace('$','',regex=True)
df['Shipped Product Sales'] = df['Ordered Product Sales'].str.replace('$','',regex=True)
```

```python
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Ordered Product Sales'] = pd.to_numeric(df['Ordered Product Sales'])
```

```python
plt.plot(df['Date'],df['Ordered Product Sales'])
plt.plot(df['Date'],df['Ordered Product Sales'],'o')
```
